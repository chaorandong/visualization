```python
# Load libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy 
import PIL
import requests
```

```python
# Sample data
np.random.seed(613)
x = np.arange(50)
y1 = np.random.randint(0, 100, 50)
y2 = np.random.randint(0, 100, 50)
```

```python
# Generate a line plot with both y variables on the same axes
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y1,y2)
```

```python
# Add a legend
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x, y1, label='Person 1')
ax.plot(x, y2, label='Person 2')
ax.legend(loc = 'lower right')
```

```python
# Modify the legend
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x, y1, label='Person 1')
ax.plot(x, y2, label='Person 2')
ax.legend(loc = 'lower right',
          frameon = True, # Add a frame around the legend
          fontsize = 12, # Change the font size
          ncol = 2, # Specify the number of columns
          shadow = True) # Add a shadow to the legend
```

```python
# Moving the legend outside of the plot area
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x, y1, label='Person 1')
ax.plot(x, y2, label='Person 2')
ax.legend(loc = 'upper left',
          bbox_to_anchor=(1, 1)) # Move the legend outside the plot area
```

```python
# Annotations added
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x, y1, label='Person 1')
ax.scatter(x, y2, label='Person 2')
ax.legend(loc = 'lower right')
ax.text(10, 95, "This value is important!")
```

```python
# Modify text annotations
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x, y1, label='Person 1')
ax.scatter(x, y2, label='Person 2')
ax.legend(loc = 'lower right')
ax.text(10, 95, "This value is important!",
        ha = 'center', ## Alignment of text
        color = 'red', ## Modify font color
        size = 20)  ## Modify font size)
```

```python
# Position text annotations
fig, ax = plt.subplots()

ax.axis([0,10,0,10])

ax.text(1, 5, ".Data:(1,5)", transform = ax.transData)
ax.text(0.5, 0.1, ".Axes:(0.5,0.1)", transform = ax.transAxes)
ax.text(0.2, 0.2, ".Figure: (0.2, 0.2)", transform=fig.transFigure)
```

```python
# Annotate with arrows
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x, y1, label='Person 1')
ax.scatter(x, y2, label='Person 2')
ax.legend(loc = 'lower right')
ax.annotate("This value is important!",
            xy=(10, 95), # Point to annotate
            xytext=(20, 94), # Text location
            #arrowprops=dict(facecolor='black', shrink=0.05)) # Arrow properties
            arrowprops = dict(arrowstyle = "wedge", color = "hotpink"))
```

```python
# Remove tick marks and labels
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x, y1, label='Person 1')
ax.scatter(x, y2, label='Person 2')
ax.legend(loc = 'lower right')
ax.yaxis.set_major_locator(plt.NullLocator())
ax.xaxis.set_major_formatter(plt.NullFormatter())
```

```python
# Limit tick marks
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x, y1, label='Person 1')
ax.scatter(x, y2, label='Person 2')
ax.legend(loc = 'lower right')
ax.xaxis.set_major_locator(plt.MaxNLocator(3)) # Limit x-axis to 3 ticks
```

```python
# Tick mark intervals
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x, y1, label='Person 1')
ax.scatter(x, y2, label='Person 2')
ax.legend(loc = 'lower right')
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
```

```python
# Rotate axis labels
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x, y1, label='Person 1')
ax.scatter(x, y2, label='Person 2')
ax.legend(loc = 'lower right')
plt.xticks(rotation=45, ha = 'right') # Rotate x-axis labels
```

```python
# Modify axis title fonts

# Define the font
font1 = {'family': 'serif', 'color': 'indigo'}

fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x, y1, label='Person 1')
ax.scatter(x, y2, label='Person 2')
ax.legend(loc = 'lower right')
plt.xlabel("Shiny New X Axis!", fontsize = 18, fontdict = font1)
```
