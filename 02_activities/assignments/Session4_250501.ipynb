Subplots and combining vis

```python
# Load libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy 
import PIL
import requests
```

```python
# Generate random data
np.random.seed(613)
x1 = np.arange(50)
y1 = np.random.randint(0,75,50)
x2 = np.array(["Luffy","Zoro","Nami","Usopp","Sanji"])
y2 = np.array([110, 180, 240, 99, 220])
```

```python
# Define two axes and their relative positions
fig, (ax1, ax2) = plt.subplots(ncols = 2, 
                               nrows=1, 
                               figsize = (7,3))
ax1.scatter(x1,y1, color = "blue")
ax2.bar(x2,y2, color = "indigo")
fig.show()
```

```python
fig, someaxes = plt.subplot_mosaic([['ax1','ax3'],
                                     ['ax2','ax3']],
                                     figsize=(7,4),
                                     layout='constrained')
someaxes['ax1'].scatter(x1,y1)
someaxes['ax2'].bar(x2,y2)
someaxes['ax3'].plot(x1,y1)
someaxes['ax1'].set_xlabel('A Big Lavel', fontsize = 18)
someaxes['ax2'].set_xlabel('Another Lable', fontsize =  18)
someaxes['ax3'].set_xlabel('Label 2: 2 Fast 2 Furious', fontsize = 18)

plt.show()
```

Multiple Viz

```python
# First make sample data
x = np.array(["Luffy","Zoro","Nami","Usopp","Sanji"])
y1 = np.array([110, 180, 240, 99, 220])
y2 = np.array([170, 100, 90, 120, 50])
```

```python
fig, ax = plt.subplots(figsize=(7,3))

ax.bar(x, y1, color = 'indigo')
ax.plot(x, y2, color = 'red')
```

```python
# Add error information
y2_sd = np.std(y2)

# Plot it
fig, ax = plt.subplots(figsize=(7,3))
ax.plot(x, y2, color = 'red')

ax.errorbar(x, y2, 
            yerr = y2_sd,
            fmt = "none",
            ecolor= "indigo",
            elinewidth= 4,
            capsize= 6,
            capthick = 4,
            errorevery=2)
```

Add images in the viz

```python
from PIL import Image
from io import BytesIO
```

```python
# Get the image from the Internet
response = requests.get("https://upload.wikimedia.org/wikipedia/en/c/cb/Monkey_D_Luffy.png")
image_file = BytesIO(response.content)
image = Image.open(image_file)
```

```python
fig, ax = plt.subplots(figsize=(7,3))
ax.plot(x, y2, color = 'red')

ax_image = fig.add_axes([0.1, # xcoordinate (on figure, not axes)
                         0.11, # y coordinate (on figure, not axes)
                         0.15, # image
                         0.35]) # image height

ax_image.imshow(image)
ax_image.axis('off') # remove the axes
```
